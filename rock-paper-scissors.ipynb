```python
import tkinter as tk
from tkinter import messagebox
import random

# Function to handle the game logic
def play(user_choice):
    choices = ['Rock', 'Paper', 'Scissors']
    computer_choice = random.choice(choices)
    user_choice_label.config(text=f"User Choice: {user_choice}")
    computer_choice_label.config(text=f"Computer Choice: {computer_choice}")
    
    if user_choice == computer_choice:
        result = "It's a Tie!"
    elif (user_choice == "Rock" and computer_choice == "Scissors") or \
         (user_choice == "Paper" and computer_choice == "Rock") or \
         (user_choice == "Scissors" and computer_choice == "Paper"):
        result = "You Win!"
        update_score('user')
    else:
        result = "Computer Wins!"
        update_score('computer')

    result_label.config(text=result)

# Function to update the score
def update_score(winner):
    global user_score, computer_score
    if winner == 'user':
        user_score += 1
        user_score_label.config(text=f"User Score: {user_score}")
    elif winner == 'computer':
        computer_score += 1
        computer_score_label.config(text=f"Computer Score: {computer_score}")

# Function to reset the game
def reset_game():
    global user_score, computer_score
    user_score = 0
    computer_score = 0
    user_score_label.config(text="User Score: 0")
    computer_score_label.config(text="Computer Score: 0")
    user_choice_label.config(text="User Choice: ")
    computer_choice_label.config(text="Computer Choice: ")
    result_label.config(text="")

# Initialize the main window
root = tk.Tk()
root.title("Rock, Paper, Scissors Game")

# Set background color
root.configure(bg="#1abc9c")

# Initialize scores
user_score = 0
computer_score = 0

# Create and place labels
title_label = tk.Label(root, text="Rock, Paper, Scissors", font=("Arial", 20), bg="#1abc9c", fg="#ffffff")
title_label.grid(row=0, column=0, columnspan=3, pady=10)

user_choice_label = tk.Label(root, text="User Choice: ", font=("Arial", 14), bg="#1abc9c", fg="#ffffff")
user_choice_label.grid(row=1, column=0, columnspan=3, pady=5)

computer_choice_label = tk.Label(root, text="Computer Choice: ", font=("Arial", 14), bg="#1abc9c", fg="#ffffff")
computer_choice_label.grid(row=2, column=0, columnspan=3, pady=5)

result_label = tk.Label(root, text="", font=("Arial", 16), bg="#1abc9c", fg="#ffffff")
result_label.grid(row=3, column=0, columnspan=3, pady=10)

user_score_label = tk.Label(root, text="User Score: 0", font=("Arial", 14), bg="#1abc9c", fg="#ffffff")
user_score_label.grid(row=4, column=0, pady=5)

computer_score_label = tk.Label(root, text="Computer Score: 0", font=("Arial", 14), bg="#1abc9c", fg="#ffffff")
computer_score_label.grid(row=4, column=2, pady=5)

# Create and place buttons
button_rock = tk.Button(root, text="Rock", command=lambda: play('Rock'), bg="#3498db", fg="#000000", font=("Arial", 12))
button_rock.grid(row=5, column=0, padx=10, pady=10)

button_paper = tk.Button(root, text="Paper", command=lambda: play('Paper'), bg="#3498db", fg="#000000", font=("Arial", 12))
button_paper.grid(row=5, column=1, padx=10, pady=10)

button_scissors = tk.Button(root, text="Scissors", command=lambda: play('Scissors'), bg="#3498db", fg="#000000", font=("Arial", 12))
button_scissors.grid(row=5, column=2, padx=10, pady=10)

# Reset button
reset_button = tk.Button(root, text="Reset", command=reset_game, bg="#e74c3c", fg="#000000", font=("Arial", 12))
reset_button.grid(row=6, column=0, columnspan=3, pady=10)

# Run the main event loop
root.mainloop()
```
